```python
import numpy as np
```

```python
# np.rand와 np.randn
# numpy의 모듈에는 ndarray를 random 값으로 초기화할 수 있는 함수들이 많이 있다.

# 예를 들어, 다음은 (균등 분포인) 0과 1사이의 random한 float point로 3x4 matrix를 초기화한다.
np.random.rand(3,4)
```

```python
# 다음은 평균이 0이고 분산이 1인 일변량 정규 분포에서 샘플링한 random한 float point를 담은 3x4 matrix이다.
np.random.randn(3,4)
```

```python
# 위 분포의 모양을 알기 위해 matplotlib 사용
%matplotlib inline
import matplotlib.pyplot as plt
plt.hist(np.random.rand(100000), density=True, bins=100, histtype="step", color="blue", label="rand")
plt.hist(np.random.randn(100000), density=True, bins=100, histtype="step", color="red", label="randn")
plt.axis([-2.5, 2.5, 0, 1.1])
plt.legend(loc = "upper left")
plt.title("Random distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```
